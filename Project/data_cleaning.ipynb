# UCDP Dataset

```python
conflict_df = pd.read_csv('datasets/ucdp-prio-acd-171.csv')
```

```python
conflict_df.head()
```

```python
for index, row in conflict_df.iterrows():
    conflict_df.loc[index, 'location'] = re.sub(r'\([^)]*\)', '', conflict_df.loc[index, 'location'])
    conflict_df.loc[index, 'location'] = conflict_df.loc[index, 'location'].rstrip()
    conflict_df.loc[index, 'sidea'] = conflict_df.loc[index, 'sidea'].replace('Government of ', '')
```

```python
conflict_df.head()
```

```python
countries_conflict_ucdp = conflict_df['location'].drop_duplicates()
countries_ucdp_without_colinizer_tmp = []

# Add all countries implicated in a conflict
for country in countries_conflict_ucdp:
    countries_ucdp_without_colinizer_tmp.extend(country.split(","))

# Remove spaces in country name
countries_ucdp_without_colinizer_tmp = [country.strip() for country in countries_ucdp_without_colinizer_tmp]
# Remove colonizer country
colonizer_countries.append('United States of America') # In a way, USA is a colonizer country... lolz
countries_ucdp_without_colinizer_tmp = [country for country in countries_ucdp_without_colinizer_tmp if country not in colonizer_countries]

# Remove duplicate
countries_ucdp_without_colinizer = set(countries_ucdp_without_colinizer_tmp)


countries_ucdp_without_colinizer
```
